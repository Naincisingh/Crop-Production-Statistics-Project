```python
## Nutrient Analysis:

## How are nutrients (N, P, K) distributed across different crops?
import matplotlib.pyplot as plt
import seaborn as sns

# Nutrient distribution across different crops
nutrients = ['N', 'P', 'K']

for nutrient in nutrients:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Crop', y=nutrient, data=df)
    plt.title(f'Distribution of {nutrient} across Different Crops')
    plt.ylabel(f'{nutrient} Levels')
    plt.xlabel('Crop Types')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
## Is there any correlation between nutrient levels and crop yield?
import pandas as pd

nutrient_yield_correlation = df[['N', 'P', 'K', 'Yield_ton_per_hec']].corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(nutrient_yield_correlation)

# Extract the correlation between nutrients and yield
correlation_N_yield = nutrient_yield_correlation.loc['N', 'Yield_ton_per_hec']
correlation_P_yield = nutrient_yield_correlation.loc['P', 'Yield_ton_per_hec']
correlation_K_yield = nutrient_yield_correlation.loc['K', 'Yield_ton_per_hec']

# Print the individual correlations
print(f"Correlation between N and Yield: {correlation_N_yield}")
print(f"Correlation between P and Yield: {correlation_P_yield}")
print(f"Correlation between K and Yield: {correlation_K_yield}")
```

## Key Insight of Nutrient Analysis :-

- Nutrient distribution varies among crops, with Potassium (K) showing the highest levels, followed by Nitrogen (N) and Phosphorus (P).
- Across different crops, no strong correlation exists between individual nutrient levels (N, P, K) and crop yield.
- The correlation matrix indicates weak positive correlations between N, P, K, and crop yield, with N having the highest correlation (0.09).
- While nutrient levels may influence yield to some extent, other factors likely contribute significantly to crop productivity.

```python
## Machine Learning Predictions:

## Build a machine learning model to predict crop yield based on the given parameters.
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# (Exclude unnecessary columns like 'Unnamed: 0', 'State_Name', 'Crop_Type', etc.)
features = df[['N', 'P', 'K', 'pH', 'rainfall', 'temperature', 'Area_in_hectares']]
target = df['Yield_ton_per_hec']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')

# Visualize actual vs. predicted values
plt.scatter(y_test, predictions)
plt.xlabel('Actual Yield (tons per hectare)')
plt.ylabel('Predicted Yield (tons per hectare)')
plt.title('Actual vs. Predicted Crop Yield')
plt.show()
```

## Key Insights :-
- The machine learning model employs a linear regression algorithm to predict crop yield based on nutrient levels (N, P, K), pH, rainfall, temperature, and area.
- Data is divided into training and testing sets (80-20 split) for model training and evaluation.
- The Mean Squared Error (MSE) is utilized as an evaluation metric, measuring the average squared difference between actual and predicted yields.
- The model achieves a MSE of 360.33, indicating the degree of variance in predicted versus actual yields.
- The scatter plot visually illustrates the correlation between actual and predicted crop yields, aiding in model performance interpretation.
